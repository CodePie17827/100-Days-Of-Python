<h2><code style="background:Gold;color:black">2.Slicing :-</code></h2>
<pre>           List slicing allows you to extract a portion (sublist) of a list by specifying 
            a range of indices.

<pre>                   <code style="background:pink;color:black"># Syntex</code> 
<pre>                           <code style="background:aqua;color:black">list[start:stop:step]
</code> 
<pre>                   
                           <code style="background:red;color:black"># Note:-</code>  
                                <code style="background:gold;color:black">1.start is the inclusive index</code>
                                <code style="background:gold;color:black">2.stop is the exclusive index</code>

```python
age=[11,12,13,18,20,25]
print(age[0:2+1])
print(age[0:3+1])
```

```python
age=[11,12,13,18,20,25]
print(age[0:5+1:1])
```

```python
age=[11,12,13,18,20,25]
print(age[0:5+1:3])
```

```python
age=[11,12,13,18,20,25]
print(age[0:5+1:4])
```

```python
character_list=["A","B","X","Y","S","V"]
selected_character=character_list[0:4]
print(selected_character)
```

```python
number_list=[1,2,3,4,5,6,7,8]
selected_portion=number_list[0:6]
print(selected_portion)
```
