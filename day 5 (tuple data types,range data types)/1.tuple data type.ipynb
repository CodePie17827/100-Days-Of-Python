```python
n=(2,3,5,88,9)

n[3]=17
print(n)
```

```python
t1=("name","harish","sawant",14,22,59)
print(t1)
print(type(t1))
```

```python
t2=(22,"vikas")
print(t2)
print(type(t2))
```

```python
#indexing in tuple
t1=("name","harish","sawant",14,22,59)
print(t1[0])
print(t1[1])
print(t1[2])
print(t1[3])
```

```python
#slicing
t1=("name","harish","sawant",14,22,59)
print(t1[0:2+1:1])
```

```python
# tuple is an immutable data type

t1=("name","harish","sawant",14,22,59)
t1[3]="vikas"
```

```python
age=(1,2,3,4,5,6,7,8)
print(age[2])
```

```python
#list
age=[11,12,13,18,20,25]
age[3]=33
print(age)
```
