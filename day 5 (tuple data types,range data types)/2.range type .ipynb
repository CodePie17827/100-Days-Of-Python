<h2><code style="background:Gold;color:black">1.Creating Range Objects::-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


                            The range data type in Python represents a sequence of numbers. You can create a 
                            range object using the range() constructor. A range can be defined with a 
                            start, stop, and step value.

```python
my_range = range(1, 10, 2)
for i in my_range:
    print(i)
```

<h2><code style="background:Gold;color:black">2.Iterating with Range:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


 


                            Range objects are commonly used in for loops to iterate over a sequence 
                            of numbers defined by the range.

```python
for i in range(5):
    print(i)
```


<h2><code style="background:Gold;color:black">
3.Converting Range to Other Data Types:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>



                            You can convert a range object to other data types like lists or tuples to 
                            manipulate or access its values more easily.

```python
my_range = range(5)
my_list = list(my_range)

print(type(my_list))
```

```python
v=range(1,20,2)
vs=str(v)
print(type(vs))
```

<h2><code style="background:Gold;color:black">4.Length of a Range:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


 



                            The length of a range data type in Python refers to the number of integer 
                            values present within the specified rang

                             You can find the length of a range using the len() function. 
                             It tells you how many values are in the range.

```python
nn=range(1,10)
print(len(nn))
```

```python
my_range = range(1, 5002, 2)
print(len(my_range))
```

```python
mmm=range(1,10000,5)
print(len(mmm))
```

<h2><code style="background:Gold;color:black">5.Indexing and Slicing:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


 




                                 Although range objects are not subscriptable, you can convert them 
                                 to other iterable types and then perform indexing and slicing operations.

```python
my_range = range(1, 10, 2)
my_list = list(my_range)
first_value = my_list[0]
print(first_value)

```

```python
#indexing
xyx=range(1,20,3)
cxyx=list(xyx)
print(cxyx[0])
print(cxyx[1])
```

```python
#indexing
xyx=range(1,20,3)
cxyx=list(xyx)
print(cxyx[0:3:1])
```

<h2><code style="background:Gold;color:black">6.Immutable Nature of Ranges:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


 




                             Ranges are immutable, meaning you cannot modify them directly. 
                             To change a range, you need to create a new one.

```python
my_range = range(1, 10, 2)
# This will raise an error
my_range[0] = 0
```

<h2><code style="background:Gold;color:black">7.
Using Range with Iteration Functions.:-</code></h2>
<pre>              <pre>                   <code style="background:pink;color:black">#Definition:-</code>


 



                         Range objects can be used as iterable sequences in various 
                         functions, such as sum(), max(), and min()

```python
my_range = range(1, 5000, 3)


print(max(my_range))
```
